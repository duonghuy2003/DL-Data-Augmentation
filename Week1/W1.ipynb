```python
import torch
import SimpleITK as sitk
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn đến file NIfTI (BraTS20_Training_001_flair.nii)
# https://www.kaggle.com/datasets/awsaf49/brats20-dataset-training-validation

file_path = "D:\Sj\PreThesis\Data\Week1\BraTS20_Training_001_flair.nii"

# Đọc ảnh bằng SimpleITK
image_sitk = sitk.ReadImage(file_path)

# Chuyển sang mảng numpy
image_np = sitk.GetArrayFromImage(image_sitk)  # Shape: (Depth, Height, Width)

print(f"Image shape: {image_np.shape}")
print(f"Image spacing: {image_sitk.GetSpacing()}")
print(f"Image origin: {image_sitk.GetOrigin()}")
```

```python
# Chọn lát cắt giữa theo trục Z (hoặc trục nào bạn muốn)
slice_idx = image_np.shape[0] // 2
slice_2d = image_np[slice_idx, :, :]

# Hiển thị lát cắt
plt.figure(figsize=(6, 6))
plt.imshow(slice_2d, cmap="gray")
plt.title(f"2D Slice at Z = {slice_idx}")
plt.axis("off")
plt.show()
```

```python
# Chọn lát cắt giữa theo trục X (sagittal)
slice_idx = image_np.shape[2] // 2  # Trục thứ 3 là width (X)
slice_2d = image_np[:, :, slice_idx]  # Lấy tất cả giá trị theo trục Z và Y

plt.figure(figsize=(6, 6))
plt.imshow(slice_2d, cmap="gray")
plt.title(f"Sagittal Slice at X = {slice_idx}")
plt.axis("off")
plt.show()
```

```python
# Chọn lát cắt giữa theo trục Y (coronal)
slice_idx = image_np.shape[1] // 2  # Trục thứ 2 là height (Y)
slice_2d = image_np[:, slice_idx, :]  # Lấy tất cả giá trị theo trục Z và X

plt.figure(figsize=(6, 6))
plt.imshow(slice_2d, cmap="gray")
plt.title(f"Coronal Slice at Y = {slice_idx}")
plt.axis("off")
plt.show()
```
