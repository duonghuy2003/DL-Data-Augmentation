```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import random
from tqdm import tqdm
import zipfile

# ================================
# 1. DATA PREPARATION
# ================================
# Unzip the dataset
DATA_ZIP = "stage1_train.zip"
EXTRACT_PATH = "stage1_train"

print("Checking for dataset...")
if os.path.exists(DATA_ZIP):
    print(f"Found {DATA_ZIP}, extracting...")
    with zipfile.ZipFile(DATA_ZIP, 'r') as zip_ref:
        zip_ref.extractall(".")
    print("Extraction complete.")
elif os.path.exists(EXTRACT_PATH):
    print(f"Dataset already exists at {EXTRACT_PATH}")
else:
    raise FileNotFoundError(f"Please upload '{DATA_ZIP}' to Colab first.")

# Count samples
sample_ids = [d for d in os.listdir(EXTRACT_PATH) if os.path.isdir(os.path.join(EXTRACT_PATH, d))]
print(f"Found {len(sample_ids)} samples in dataset")

# Use GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# ================================
# 2. STRICT U-NET ARCHITECTURE (Ronneberger et al. 2015)
# ================================
class DoubleConv(nn.Module):
    """(convolution => [BN] => ReLU) * 2"""
    def __init__(self, in_channels, out_channels, mid_channels=None):
        super().__init__()
        if not mid_channels:
            mid_channels = out_channels
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, mid_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(mid_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(mid_channels, out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)


class Down(nn.Module):
    """Downscaling with maxpool then double conv"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.maxpool_conv = nn.Sequential(
            nn.MaxPool2d(2),
            DoubleConv(in_channels, out_channels)
        )

    def forward(self, x):
        return self.maxpool_conv(x)


class Up(nn.Module):
    """Upscaling then double conv"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.up = nn.ConvTranspose2d(in_channels, in_channels // 2, kernel_size=2, stride=2)
        self.conv = DoubleConv(in_channels, out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)
        # Pad if necessary
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]

        x1 = F.pad(x1, [diffX // 2, diffX - diffX // 2,
                        diffY // 2, diffY - diffY // 2])
        x = torch.cat([x2, x1], dim=1)
        return self.conv(x)


class UNet(nn.Module):
    """Original U-Net Architecture (Ronneberger et al. 2015)"""
    def __init__(self, n_channels=3, n_classes=1):
        super(UNet, self).__init__()
        self.n_channels = n_channels
        self.n_classes = n_classes

        # Encoder (Contracting Path)
        self.inc = DoubleConv(n_channels, 64)      # 64 filters
        self.down1 = Down(64, 128)                 # 128 filters
        self.down2 = Down(128, 256)                # 256 filters
        self.down3 = Down(256, 512)                # 512 filters
        self.down4 = Down(512, 1024)               # 1024 filters (bottleneck)

        # Decoder (Expansive Path)
        self.up1 = Up(1024, 512)                   # 1024 -> 512
        self.up2 = Up(512, 256)                    # 512 -> 256
        self.up3 = Up(256, 128)                    # 256 -> 128
        self.up4 = Up(128, 64)                     # 128 -> 64

        # Output layer
        self.outc = nn.Conv2d(64, n_classes, kernel_size=1)

    def forward(self, x):
        # Encoder
        x1 = self.inc(x)      # 64 channels
        x2 = self.down1(x1)   # 128 channels
        x3 = self.down2(x2)   # 256 channels
        x4 = self.down3(x3)   # 512 channels
        x5 = self.down4(x4)   # 1024 channels (bottleneck)

        # Decoder with skip connections
        x = self.up1(x5, x4)  # 1024 -> 512 (concatenates with x4)
        x = self.up2(x, x3)   # 512 -> 256 (concatenates with x3)
        x = self.up3(x, x2)   # 256 -> 128 (concatenates with x2)
        x = self.up4(x, x1)   # 128 -> 64 (concatenates with x1)

        # Output
        logits = self.outc(x)
        return logits


# ================================
# 3. DATASET CLASS
# ================================
class NucleiDataset(Dataset):
    def __init__(self, root_dir, img_size=256, train=True):
        self.root_dir = root_dir
        self.img_size = img_size
        self.train = train

        # Get all valid image IDs
        self.image_ids = []
        for img_id in os.listdir(root_dir):
            img_path = os.path.join(root_dir, img_id, "images", f"{img_id}.png")
            mask_dir = os.path.join(root_dir, img_id, "masks")
            if os.path.exists(img_path) and os.path.exists(mask_dir):
                self.image_ids.append(img_id)

        print(f"Dataset loaded: {len(self.image_ids)} samples")

        # Image normalization (ImageNet stats)
        self.normalize = transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        img_id = self.image_ids[idx]

        # Load image
        img_path = os.path.join(self.root_dir, img_id, "images", f"{img_id}.png")
        image = Image.open(img_path).convert("RGB")

        # Load and combine masks
        mask_dir = os.path.join(self.root_dir, img_id, "masks")
        mask = None
        for mask_file in os.listdir(mask_dir):
            mask_path = os.path.join(mask_dir, mask_file)
            m = np.array(Image.open(mask_path).convert("L"))
            if mask is None:
                mask = m
            else:
                mask = np.maximum(mask, m)

        # Resize
        image = image.resize((self.img_size, self.img_size), Image.BILINEAR)
        mask = Image.fromarray(mask).resize((self.img_size, self.img_size), Image.NEAREST)

        # Convert to tensors
        image_tensor = transforms.ToTensor()(image)
        mask_tensor = transforms.ToTensor()(mask)

        # Apply augmentations if training
        if self.train and random.random() > 0.5:
            # Horizontal flip
            image_tensor = transforms.functional.hflip(image_tensor)
            mask_tensor = transforms.functional.hflip(mask_tensor)

        if self.train and random.random() > 0.5:
            # Vertical flip
            image_tensor = transforms.functional.vflip(image_tensor)
            mask_tensor = transforms.functional.vflip(mask_tensor)

        # Normalize image
        image_tensor = self.normalize(image_tensor)

        # Binary mask (0 or 1)
        mask_tensor = (mask_tensor > 0.5).float()

        return image_tensor, mask_tensor


# ================================
# 4. METRICS CALCULATION
# ================================

def calculate_dice(pred, target, smooth=1e-8):
    """Calculate Dice coefficient for a single image or batch"""
    # If pred is still logits (unactivated), apply sigmoid. 
    # If it's already a mask (0-1), this won't hurt much, but it's better to be explicit.
    if pred.min() < 0 or pred.max() > 1:
        pred = torch.sigmoid(pred)
    
    pred = (pred > 0.5).float()
    target = (target > 0.5).float()

    # Flatten tensors to ensure we are comparing pixel-to-pixel
    pred_flat = pred.view(-1)
    target_flat = target.view(-1)

    intersection = (pred_flat * target_flat).sum()
    dice = (2. * intersection + smooth) / (pred_flat.sum() + target_flat.sum() + smooth)
    
    return dice.item()

def calculate_iou(pred, target, smooth=1e-8):
    """Calculate Intersection over Union"""
    if pred.min() < 0 or pred.max() > 1:
        pred = torch.sigmoid(pred)
        
    pred = (pred > 0.5).float()
    target = (target > 0.5).float()

    pred_flat = pred.view(-1)
    target_flat = target.view(-1)

    intersection = (pred_flat * target_flat).sum()
    union = pred_flat.sum() + target_flat.sum() - intersection
    
    iou = (intersection + smooth) / (union + smooth)
    return iou.item()


# ================================
# 5. TRAINING FUNCTION
# ================================
def train_model():
    """Main training function"""
    print("\n" + "="*60)
    print("U-NET TRAINING - DATA SCIENCE BOWL 2018")
    print("="*60)

    # Hyperparameters
    BATCH_SIZE = 8
    LEARNING_RATE = 1e-4
    EPOCHS = 50
    PATIENCE = 10  # Early stopping patience
    IMG_SIZE = 256

    # Create dataset and split
    print("\n1. Preparing dataset...")
    full_dataset = NucleiDataset(EXTRACT_PATH, img_size=IMG_SIZE, train=True)

    # Split dataset (85% train, 15% validation)
    train_size = int(0.85 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    print(f"   Total samples: {len(full_dataset)}")
    print(f"   Training: {len(train_dataset)}")
    print(f"   Validation: {len(val_dataset)}")

    # Disable augmentation for validation
    val_dataset.dataset.train = False

    # Create data loaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=BATCH_SIZE,
        shuffle=True,
        num_workers=2,
        pin_memory=True  # Faster data transfer to GPU
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=BATCH_SIZE,
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )

    # Initialize model
    print("\n2. Initializing U-Net model...")
    model = UNet(n_channels=3, n_classes=1).to(device)

    # Count parameters
    total_params = sum(p.numel() for p in model.parameters())
    print(f"   Total parameters: {total_params:,}")
    print(f"   Model structure:")
    print(f"     Input: 3 channels")
    print(f"     Encoder: 64 → 128 → 256 → 512 → 1024")
    print(f"     Decoder: 1024 → 512 → 256 → 128 → 64")
    print(f"     Output: 1 channel")

    # Loss function and optimizer
    criterion = nn.BCEWithLogitsLoss()  # Binary cross-entropy with logits
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)

    # Learning rate scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=0.5, patience=5
    )

    print("\n3. Starting training...")
    best_dice = 0.0
    no_improve = 0
    history = {'train_loss': [], 'val_dice': [], 'val_iou': []}

    for epoch in range(EPOCHS):
        # Training phase
        model.train()
        train_loss = 0.0

        for images, masks in tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}", unit="batch"):
            images = images.to(device)
            masks = masks.to(device)

            # Forward pass
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, masks)

            # Backward pass
            loss.backward()
            optimizer.step()

            train_loss += loss.item()

        avg_train_loss = train_loss / len(train_loader)
        history['train_loss'].append(avg_train_loss)

        # Validation phase
        model.eval()
        val_dice_scores = []
        val_iou_scores = []

        with torch.no_grad():
            for images, masks in val_loader:
                images = images.to(device)
                masks = masks.to(device)

                outputs = model(images)

                dice = calculate_dice(outputs, masks)
                iou = calculate_iou(outputs, masks)

                val_dice_scores.append(dice)
                val_iou_scores.append(iou)

        avg_val_dice = np.mean(val_dice_scores)
        avg_val_iou = np.mean(val_iou_scores)
        history['val_dice'].append(avg_val_dice)
        history['val_iou'].append(avg_val_iou)

        # Update learning rate
        scheduler.step(avg_val_dice)

        # Print epoch results
        print(f"   Train Loss: {avg_train_loss:.4f}")
        print(f"   Val Dice:   {avg_val_dice:.4f}")
        print(f"   Val IoU:    {avg_val_iou:.4f}")

        # Save best model
        if avg_val_dice > best_dice:
            best_dice = avg_val_dice
            no_improve = 0
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'val_dice': avg_val_dice,
                'val_iou': avg_val_iou,
            }, 'best_unet_model.pth')
            print(f"   ✓ Saved best model (Dice: {best_dice:.4f})")
        else:
            no_improve += 1
            if no_improve >= PATIENCE:
                print(f"\n   Early stopping triggered!")
                break

        print("-" * 50)

    print("\n4. Training completed!")
    print(f"   Best validation Dice: {best_dice:.4f}")

    return model, history


# ================================
# 6. VISUALIZATION FUNCTIONS
# ================================
def plot_training_history(history):
    """Plot training curves"""
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    epochs = range(1, len(history['train_loss']) + 1)

    # Loss plot
    axes[0].plot(epochs, history['train_loss'], 'b-', linewidth=2)
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].set_title('Training Loss')
    axes[0].grid(True, alpha=0.3)

    # Dice plot
    axes[1].plot(epochs, history['val_dice'], 'g-', linewidth=2)
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Dice Score')
    axes[1].set_title('Validation Dice Score')
    axes[1].grid(True, alpha=0.3)
    axes[1].set_ylim([0, 1])

    # IoU plot
    axes[2].plot(epochs, history['val_iou'], 'r-', linewidth=2)
    axes[2].set_xlabel('Epoch')
    axes[2].set_ylabel('IoU Score')
    axes[2].set_title('Validation IoU Score')
    axes[2].grid(True, alpha=0.3)
    axes[2].set_ylim([0, 1])

    plt.tight_layout()
    plt.show()


def visualize_predictions(model, dataset, num_samples=3):
    """Visualize model predictions"""
    model.eval()

    # Randomly select samples
    indices = random.sample(range(len(dataset)), num_samples)

    fig, axes = plt.subplots(num_samples, 4, figsize=(16, 4 * num_samples))

    for i, idx in enumerate(indices):
        image, true_mask = dataset[idx]

        # Add batch dimension and move to device
        image_batch = image.unsqueeze(0).to(device)

        # Get prediction
        with torch.no_grad():
            output = model(image_batch)
            pred_mask = torch.sigmoid(output).squeeze().cpu().numpy()

        # Convert to binary mask
        pred_binary = (pred_mask > 0.5).astype(np.float32)

        # Denormalize image for display
        image_np = image.numpy().transpose(1, 2, 0)
        image_np = image_np * [0.229, 0.224, 0.225] + [0.485, 0.456, 0.406]
        image_np = np.clip(image_np, 0, 1)

        # True mask
        true_mask_np = true_mask.squeeze().numpy()

        # Calculate metrics
        dice = calculate_dice(
            torch.tensor(pred_mask).unsqueeze(0).unsqueeze(0),
            true_mask.unsqueeze(0)
        )

        iou = calculate_iou(
            torch.tensor(pred_mask).unsqueeze(0).unsqueeze(0),
            true_mask.unsqueeze(0)
        )

        # Plot
        ax_row = axes[i] if num_samples > 1 else axes

        ax_row[0].imshow(image_np)
        ax_row[0].set_title("Input Image")
        ax_row[0].axis('off')

        ax_row[1].imshow(true_mask_np, cmap='gray')
        ax_row[1].set_title("Ground Truth")
        ax_row[1].axis('off')

        ax_row[2].imshow(pred_binary, cmap='gray')
        ax_row[2].set_title(f"Prediction\nDice: {dice:.3f}, IoU: {iou:.3f}")
        ax_row[2].axis('off')

        ax_row[3].imshow(image_np)
        ax_row[3].imshow(pred_binary, alpha=0.5, cmap='jet')
        ax_row[3].set_title("Overlay")
        ax_row[3].axis('off')

    plt.tight_layout()
    plt.show()


# ================================
# 7. MAIN EXECUTION
# ================================
if __name__ == "__main__":
    try:
        # Train the model
        model, history = train_model()

        # Plot training history
        print("\n5. Plotting training history...")
        plot_training_history(history)

        # Load best model for visualization
        print("\n6. Loading best model for visualization...")
        checkpoint = torch.load('best_unet_model.pth', map_location=device)
        model.load_state_dict(checkpoint['model_state_dict'])
        print(f"   Best model loaded (Dice: {checkpoint['val_dice']:.4f})")

        # Create validation dataset for visualization
        val_dataset = NucleiDataset(EXTRACT_PATH, img_size=256, train=False)

        # Visualize predictions
        print("\n7. Visualizing predictions...")
        visualize_predictions(model, val_dataset, num_samples=4)


        print("\n" + "="*60)
        print("TRAINING COMPLETE!")
        print("="*60)

    except Exception as e:
        print(f"\nError: {e}")
        import traceback
        traceback.print_exc()
```
